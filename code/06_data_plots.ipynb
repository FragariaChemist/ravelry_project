```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Code is from seaborn documentation.  Experimenting with creating general plot setting for the entire notebook
sns.set_theme(context='notebook', style='darkgrid', palette='Set2', font='sans-serif', font_scale=1, color_codes=True, rc=None)
```

```python
patterns_df = pd.read_csv('../rav_clean.csv'
```

```python
patterns_df
```

```python
author_group_df = pd.DataFrame(patterns_df.groupby(['author'])['type'].value_counts())
```

```python
top_author_group_df = author_group_df.sort_values(by = 'count', ascending = False).head(20)
```

```python
plt.figure(figsize = (12,8))

sns.barplot(data = top_author_group_df,
            x = 'count',
            y = 'author',
            hue = 'type',
)

plt.xticks(fontsize = 25)
plt.yticks(fontsize = 20)
plt.xlabel('Number of Patterns', fontsize = 30)
plt.ylabel('Author', fontsize = 30)
plt.title('Patterns per Author - By Type', fontsize = 35)
plt.legend(fontsize=25)

plt.tight_layout()
plt.savefig('../plot_images/patterns_by_author_type.png', dpi = 192);
```
